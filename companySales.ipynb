```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("csd.csv")
pd.set_option('display.max_columns', None)
df.head()
```

```python
# check errors in dataset
df.info()
# if missing values do df.fillna(0)
```

```python
#df.describe
df.corr()
```

```python
# finding corelation
#print(df.corr())
import seaborn as sn
sn.heatmap(df.corr(), cmap="YlGnBu", annot=True)
plt.show()
```

```python
#Read Total profit of all months and show it using a line plot.
# prifit = sp - cp

profit = df['total_profit'].tolist()
month = df['month_number'].tolist()
plt.plot(month, profit)
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(month)
#plt.yticks(profit)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```

```python
# Get total profit of all months and show line plot with the following Style properties
profit = df['total_profit'].tolist()
month = df['month_number'].tolist()
plt.plot(month, profit, linestyle = 'dashed')
plt.xlabel('Month number')
plt.ylabel('Profit in dollar')
plt.xticks(month)
#plt.yticks(profit)
plt.title('Company profit per month')
plt.yticks([100000, 200000, 300000, 400000, 500000])
plt.show()
```
